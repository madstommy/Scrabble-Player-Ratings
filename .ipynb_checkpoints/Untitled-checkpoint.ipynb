```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
games_raw = pd.read_csv('data/games.csv')
turns_raw = pd.read_csv('data/turns.csv')
train_raw = pd.read_csv('data/train.csv')
test_raw = pd.read_csv('data/test.csv')
games_raw.head()
```

```python
turns_raw.head()
```

```python
train_raw['nickname'].value_counts()
```

```python
human_players = train_raw[ ~train_raw['nickname'].isin(['BetterBot', 'STEEBot', 'HastyBot']) ]
train_y = human_players['rating']
train_x = human_players.drop(columns='rating')

test_human_players = test_raw[ ~test_raw['nickname'].isin(['BetterBot', 'STEEBot', 'HastyBot']) ]
test_x = test_human_players.drop(columns='rating')
```

```python
######### Transformations to games.csv ##########
def human_first(nickname):
    if nickname in ['BetterBot', 'STEEBot', 'HastyBot']:
        return 0
    else:
        return 1
```

```python
games_data = pd.DataFrame()
games_data['game_id'] = games_raw['game_id']
games_data['human_first'] = games_raw['first'].apply(human_first)
games_data['time_control'] = games_raw['time_control_name'].factorize()[0]
games_data['game_end_reason'] = games_raw['game_end_reason'].factorize()[0]
games_data['winner'] = games_raw['winner']
games_data['lexicon'] = games_raw['lexicon'].factorize()[0]
games_data['increment_seconds'] = games_raw['increment_seconds']
games_data['initial_time_seconds'] = games_raw['initial_time_seconds']
games_data['ranked'] = games_raw['rating_mode'].factorize()[0]
games_data['max_overtime_minutes'] = games_raw['max_overtime_minutes']
games_data['game_duration_seconds'] = games_raw['game_duration_seconds']
games_data.head()
```

```python
train_x_2 = train_x.merge(games_data, how='inner', on='game_id')
test_x_2 = test_x.merge(games_data, how='inner', on='game_id')
test_x_2.info()
```

```python
######### Transform Turns.csv, need to aggregate the data to match up the rows 

turns_raw.head()
```

```python
average_points = turns_raw.groupby(by=['game_id', 'nickname']).mean()
average_points = average_points.reset_index()
human_points = average_points[ ~average_points['nickname'].isin(['BetterBot', 'STEEBot', 'HastyBot']) ]
train_x_final = train_x_2.merge(human_points, how='inner', on='game_id')

test_x_final = test_x_2.merge(human_points, how='inner', on='game_id')
test_x_final.head()
```

```python
train_x_final.head()
train_x_final = train_x_final.drop(columns=['game_id', 'nickname_x', 'score_y', 'nickname_y', 'turn_number'])
test_ids = test_x_final['game_id']
test_x_final = test_x_final.drop(columns=['game_id', 'nickname_x', 'score_y', 'nickname_y', 'turn_number'])
```

```python
train_data = train_x_final
train_data['rating'] = train_y.reset_index()['rating']
train_data.head()
```

```python
plt.style.use('ggplot')
```

```python
sns.heatmap(data=train_data.corr())
```

```python
fig, axs = plt.subplots(nrows=4, ncols=3, figsize=(20,20), dpi=200)

for i in range(len(axs.flatten())):
    sns.scatterplot(data=train_data, x=train_data.columns[i], y='rating', linewidth=0 , alpha=0.1, ax=axs.flatten()[i])
fig.tight_layout()
```

```python
cols_to_drop = ['points', 'human_first', 'time_control', 'game_end_reason', 'winner']
train_x_final = train_x_final.drop(columns=cols_to_drop)
test_x_final = test_x_final.drop(columns=cols_to_drop)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='error', drop='first')
enc.fit(train_x_final[['lexicon']])
t = enc.transform(train_x_final[['lexicon']]).toarray()

train_x_final['lexicon_1'] = t[:, 0]
train_x_final['lexicon_2'] = t[:, 1]
train_x_final['lexicon_3'] = t[:, 2]
train_x_final = train_x_final.drop(columns='lexicon')
```

```python
f = enc.transform(test_x_final[['lexicon']]).toarray()

test_x_final['lexicon_1'] = f[:, 0]
test_x_final['lexicon_2'] = f[:, 1]
test_x_final['lexicon_3'] = f[:, 2]
test_x_final = test_x_final.drop(columns='lexicon')
```

```python
train_x_final = train_x_final.drop(columns='rating')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = dict()
gs = GridSearchCV(LinearRegression(), param_grid, scoring='neg_root_mean_squared_error', cv=10)
gs.fit(train_x_final, train_y)
est = gs.best_estimator_
print(gs.best_score_ * -1) 
```

```python
results = est.predict(test_x_final)
submission = pd.DataFrame()
submission['game_id'] = test_ids
submission['rating'] = results
submission.to_csv('data/submission.csv', index=False)
```
